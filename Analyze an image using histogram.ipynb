```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_w = np.ones((600,700,3), np.uint8)*255
blank_b = np.zeros((600,700,3), np.uint8)*255
```

```python
#hist = cv2.calcHist([blank_w], [0], None, [256], [0,256]) 
hist = cv2.calcHist([blank_b], [0], None, [256], [0,256])

plt.plot(hist)

plt.show()

cv2.imshow("blank w", blank_w) 
cv2.imshow("blank b", blank_b)

cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))

img1 = cv2.imread("D.jpeg")

img2 = cv2.imread("D1.jpeg")

img3 = cv2.imread("D2.jpeg")

hist1 = cv2.calcHist([img1], [0], None, [256], [0,256])

hist2 = cv2.calcHist([img2], [0], None, [256], [0,256])

hist3 = cv2.calcHist([img3], [0], None, [256], [0,256])

plt.subplot(3,1,1) 
plt.plot(hist1)

plt.subplot(3,1,2) 
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()

cv2.imshow("blank bl", img1) 
cv2.imshow("blank b2", img2) 
cv2.imshow("blank b3", img3)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,10))

img1 = cv2.imread("SH.jpg")
img1 = cv2.resize(img1, (600,400))

b,g,r = cv2.split(img1)

hist1 = cv2.calcHist([b], [0], None, [256], [0,256])

hist2 = cv2.calcHist([g], [0], None, [256], [0,256])

hist3 = cv2.calcHist([r], [0], None, [256], [0,256])

plt.subplot(3,1,1) 
plt.plot(hist1)

plt.subplot(3,1,2) 
plt.plot(hist2)

plt.subplot(3,1,3)
plt.plot(hist3)

plt.show()

cv2.imshow("blank bl", img1) 
cv2.imshow("blank b", b)
cv2.imshow("blank g", g) 
cv2.imshow("blank r", r)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
#plt.figure(figsize=(8,8))

img1 = cv2.imread("SH.jpg")
img1 = cv2.resize(img1, (600,400))

b,g,r = cv2.split(img1)

plt.hist(b.ravel(),255,(0,255),color="blue")
plt.hist(g.ravel(),255,(0,255),color="green")
plt.hist(r.ravel(),255,(0,255),color="red")

plt.show()

cv2.imshow("blank bl", img1) 
cv2.imshow("blank b", b)
cv2.imshow("blank g", g) 
cv2.imshow("blank r", r)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```
