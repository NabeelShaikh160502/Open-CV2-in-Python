```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("Leaf.png") 
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

hist = cv2.calcHist([img], [0], None, [255], [0,255])

plt.plot(hist)
plt.show()

cv2.imshow("wscube", img) 
cv2.waitKey(0)
cv2.destroyAllWindows()
```
